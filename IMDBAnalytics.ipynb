```python
import pandas as pd
from numpy import genfromtxt
import numpy as np
hyper = genfromtxt('hyper4.csv', delimiter=',')

bayes = genfromtxt('bayes4.csv', delimiter=',')

ax = genfromtxt('ax4.csv', delimiter=',')

ng = genfromtxt('ng4.csv', delimiter=',')

bohb = genfromtxt('bohb4.csv', delimiter=',')

ran = genfromtxt('ran4.csv', delimiter=',')

zoo =  genfromtxt('zoo4.csv', delimiter=',')

def san(a):
    t = 0
    for j in range(4):
        a[:,2*j] = a[:,2*j+t,]
        t+=1
        a[:,2*j+1] = a[:,2*j+t+1]

        
def maxof(a):
    for j in range(4):
        ma = 10000;
        ta = 10000;
        for i in range(128):
            if(ma>a[i,2*j+1]):
                ta = a[i,2*j]
            ma = min(ma,a[i,2*j+1]) 
            a[i,2*j+1] = ma
            a[i,2*j] = ta

        
def all(x):
    san(x)
    maxof(x)
    return x[:,0:8]
hyper = all(hyper)
bayes = all(bayes)
ng = all(ng)
ax = all(ax)
bohb = all(bohb)
ran = all(ran)
zoo = all(zoo)

        
        
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline
import matplotlib.pyplot as plt

def getall(a):
    f = [b for b in a[:,::2].mean(1)]
    g = [b for b in a[:,1::2].mean(1)]
    return (f, a[:,::2].std(1)/2, g,
                a[:,1::2].std(1)/2)

        
```
