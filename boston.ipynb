```python
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline
import matplotlib.pyplot as plt

def maxof(a):
    for j in range(4):
        ma = 10000;
        ta = 10000;
        for i in range(256):
            if(ma>a[i,2*j+1]):
                ta = a[i,2*j]
            ma = min(ma,a[i,2*j+1]) 
            a[i,2*j+1] = ma
            a[i,2*j] = ta

def getall(a):
    f = [min(b,0.02) for b in a[:,::2].mean(1)]
    g = [min(b,0.02) for b in a[:,1::2].mean(1)]
    return (f, a[:,::2].std(1)/2, g,
                a[:,1::2].std(1)/2)
def getallI(a):
    t = np.arange(256)

    return (InterpolatedUnivariateSpline(t,a[:,::2].mean(1))(t), 
            InterpolatedUnivariateSpline(t,a[:,::2].std(1))(t)/4, 
            InterpolatedUnivariateSpline(t,a[:,1::2].mean(1))(t),
            InterpolatedUnivariateSpline(t, a[:,1::2].std(1))(t)/4)
           
   

```
