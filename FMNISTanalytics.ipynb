```python
import pandas as pd
from numpy import genfromtxt
import numpy as np
hyper = genfromtxt('hyper3.csv', delimiter=',')

bayes = genfromtxt('bayes3.csv', delimiter=',')

ax = genfromtxt('ax3.csv', delimiter=',')

ng = genfromtxt('ng3.csv', delimiter=',')

bohb = genfromtxt('bohb3.csv', delimiter=',')

ran = genfromtxt('ran3.csv', delimiter=',')

zoo =  genfromtxt('zoo3.csv', delimiter=',')


def san(a):
    t = 0
    for j in range(5-(a.shape==bohb.shape)):
        a[:,2*j] = a[:,2*j+t,]
        t+=1
        a[:,2*j+1] = a[:,2*j+t+1]
        
        
def minof(a):
    for j in range(4):
        ma = 0;
        ta = 0;
        for i in range(128):
            if(ma<a[i,0+j*2]):
                ta = a[i,1+j*2]
            ma = max(ma,a[i,0+j*2]) 
            a[i,0+j*2] = ma
            a[i,1+j*2] = ta
        
def all(x):
    minof(x)
   
    return x #[:,0:10]
        
hyper = all(hyper)
bayes = all(bayes)
ng = all(ng)
ax = all(ax)
bohb = all(bohb[::-1,])
ran = all(ran)


correctzoopt(zoo)
zoo = all(zoo)



def correctzoopt(x):
    for j in range(4):
        x[:,2*j] = 1-x[:,2*j]
        
        
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline
import matplotlib.pyplot as plt

def maxof(a):
    for j in range(4):
        ma = 10000;
        ta = 10000;
        for i in range(256):
            if(ma>a[i,2*j+1]):
                ta = a[i,2*j]
            ma = min(ma,a[i,2*j+1]) 
            a[i,2*j+1] = ma
            a[i,2*j] = ta

def getall(a):
    f = [b for b in a[:,1::2].mean(1)]
    g = [b for b in a[:,0::2].mean(1)]
    return (f, a[:,1::2].std(1)/2, g,
                a[:,0::2].std(1)/2)
def getallI(a):
    t = np.arange(256)

    return (InterpolatedUnivariateSpline(t,a[:,1::2].mean(1))(t), 
            InterpolatedUnivariateSpline(t,a[:,1::2].std(1))(t)/4, 
            InterpolatedUnivariateSpline(t,a[:,0::2].mean(1))(t),
            InterpolatedUnivariateSpline(t, a[:,0::2].std(1))(t)/4)
           


        
```
