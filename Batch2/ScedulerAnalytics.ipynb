```python
def splitl(x,a,b):
    n=7500
    c=np.array(x[0:a])
    d=np.array(x[a:b])
    f=np.array(x[b:])
    return c,d,f
```

```python
def split(x,a,b):
    n=7500
    c=np.array(x[0:a])
    d=np.array(x[a:b])
    f=np.array(x[b:])
    c.resize(n)
    d.resize(n)
    f.resize(n)
    return np.array((c+np.zeros((n)),d
                     +np.zeros((n)), f+np.zeros((n))))
```
